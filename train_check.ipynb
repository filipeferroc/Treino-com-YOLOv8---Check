```python
import shutil
import os

source = '/content/runs'
destination = '/content/drive/MyDrive/Faculdade/TCC/Treinamento_Yolo/Treino_1'

# Copiar o conteúdo de cada subdiretório e arquivo
for item in os.listdir(source):
    s = os.path.join(source, item)
    d = os.path.join(destination, item)
    if os.path.isdir(s):
        shutil.copytree(s, d, dirs_exist_ok=True)  # Requer Python 3.8 ou superior
    else:
        shutil.copy2(s, d)



```

```python
model = YOLO('/content/drive/MyDrive/Faculdade/TCC/Treinamento_Yolo/Treino_1/detect/train/weights/best.pt')
```

```python
# Caminho da imagem para realizar a predição
image_path = '/content/drive/MyDrive/Faculdade/TCC/Imagens/prova_2_modo_2.jpg'

# Fazer a predição
results = model.predict(source=image_path, save=True)
```

```python
for result in results:
    print(result.boxes.xyxy)  # Coordenadas das caixas delimitadoras (x1, y1, x2, y2)
    print(result.boxes.conf)  # Confiança das predições
    print(result.boxes.cls)   # Classes previstas (índices das classes)
```
